A continuacion importo nueva tabal con el iata code y el nombre de la ciudad


```python
datas = pd.read_json('https://raw.githubusercontent.com/mwgg/Airports/master/airports.json') 
datas.head()
```

Como quiero hacer un diccionario de las columnas (pd series) hago transpose

```python
datas = datas.T
datas.head()
```

De esta tabla voy a sacar un diccionario para luego mapear con el codigo iata y la ciudad


```python

dicc = pd.Series(datas.city.values,index=datas.iata).to_dict()

dicc
```

Una vez que ya lo tengo me hago un df llamado bookings con las columnas que me interesan asio tarda menos en ejecutarse las cosas

```python
book.head()
```

```python
bookings = book[['iata', 'pax']]
```

```python
bookings.head()
```

inclumos una columna con la ciudad, haciendo mapping con el diccionario que hemos creado

```python
bookings['city'] = bookings['iata'].map(dicc) 
```

```python
bookings.head()
```

creamos un groupby por aeropuerto sumando pax

```python
by_arr = bookings.groupby('iata').sum()
```

Ahora lo agrupamos de mayor a menor

```python
by_arr = by_arr.sort_values(by = 'pax',ascending = False)
```

```python
by_arr
```

Realizamos un reset index para hacer el maping de nuevo con la ciudad

```python
by_arr_city = by_arr.reset_index()
```

```python
by_arr_city['city'] = by_arr_city['iata'].map(dicc)
```

Ponemos el indice de nuevo

```python
by_arr_city.set_index('iata', inplace=True)
```

Simplemente ponemos el top 10 y listo

```python
by_arr_city.head(10)
```
